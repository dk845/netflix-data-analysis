 Feature 1: User Recommendation System (Content-Based)

 What this cell does:
Uses the metadata (title, genre, description) to recommend similar shows

Uses TfidfVectorizer to encode text

Uses cosine similarity to recommend titles based on content


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Fill NaNs and combine useful metadata into one string
df['combined'] = df['type'].fillna('') + ' ' + df['listed_in'].fillna('') + ' ' + df['description'].fillna('')

df['combined'] = df['combined'].str.lower()
```

```python
# Vectorize the combined text
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['combined'])

# Compute cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Reset index for future use
df = df.reset_index()
indices = pd.Series(df.index, index=df['title']).drop_duplicates()

def recommend(title, cosine_sim=cosine_sim):
    idx = indices.get(title)
    if idx is None:
        return "Title not found in dataset."
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:6]
    show_indices = [i[0] for i in sim_scores]
    return df['title'].iloc[show_indices]
```

```python
# Try giving any title that exists in your dataset
recommend("Narcos")
```

 Feature 2: Time Series Genre Trends Over Years


 What this cell does:
Analyzes how the distribution of genres has evolved over the years.

Creates a time series plot for each genres popularity.

Allows users to see which genres gained or lost traction.

```python
# Convert 'release_year' to numeric
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

# Drop rows where release_year or genre is missing
df = df.dropna(subset=['release_year', 'listed_in'])

# Extract the genres from 'listed_in' column
df['genres'] = df['listed_in'].str.split(', ')

# Flatten the DataFrame into multiple rows for each genre
df_exploded = df.explode('genres')

# Ensure the data is sorted by year
df_exploded = df_exploded[df_exploded['release_year'] >= 2000]  # Filter years after 2000
df_exploded.head()
```

```python
# Count occurrences of each genre per year
genre_year_counts = df_exploded.groupby(['release_year', 'genres']).size().reset_index(name='count')

# Create a pivot table to have years as columns and genres as rows
genre_pivot = genre_year_counts.pivot(index='genres', columns='release_year', values='count').fillna(0)

# Display the pivot table to get an idea of genre trends
genre_pivot.head()
```

```python
# Plot genre trends over the years
plt.figure(figsize=(14, 8))
sns.lineplot(data=genre_pivot.T, dashes=False, palette='tab20')
plt.title('Genre Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.legend(title='Genres', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

 Explanation:

The explosion (df.explode('genres')) step makes each genre into its own row, which helps analyze each genre's presence year by year.

The pivot table reshapes the data, where each genre has its counts across years.

The line plot visually presents how each genre's presence evolved over the years.



 Feature 3: Sentiment Analysis of Descriptions

 What this cell does:

Analyzes the sentiment of the movie or show descriptions to see whether the tone is more positive, neutral, or negative.

Uses a simple sentiment analysis model (VADER, a lexicon and rule-based sentiment analysis tool).

Helps understand whether Netflix's content descriptions tend to be positive or negative over time.

```python
!pip install nltk
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the necessary NLTK data
nltk.download('vader_lexicon')
```

```python
# Initialize Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()

# Define a function to get sentiment scores
def get_sentiment(description):
    sentiment_score = sia.polarity_scores(description)
    return sentiment_score['compound']  # We use the compound score for overall sentiment

# Apply the function to the 'description' column
df['sentiment_score'] = df['description'].apply(get_sentiment)

# Show the sentiment score for a few rows
df[['description', 'sentiment_score']].head()
```

```python
# Plot sentiment score distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['sentiment_score'], kde=True, color='blue', bins=30)
plt.title('Sentiment Distribution of Netflix Descriptions')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Calculate the average sentiment per year
avg_sentiment_by_year = df.groupby('release_year')['sentiment_score'].mean()

# Plot the trend of average sentiment over the years
plt.figure(figsize=(12, 6))
sns.lineplot(x=avg_sentiment_by_year.index, y=avg_sentiment_by_year.values, color='green')
plt.title('Average Sentiment of Netflix Descriptions Over Years')
plt.xlabel('Year')
plt.ylabel('Average Sentiment Score')
plt.grid(True)
plt.show()
```

 Explanation:

Sentiment Analysis: The SentimentIntensityAnalyzer from VADER assigns a compound score (ranging from -1 to 1) based on the tone of the description, where positive values indicate positive sentiment and negative values indicate negative sentiment.

Distribution Plot: The histogram helps visualize how the sentiment of descriptions is distributed across positive, neutral, and negative sentiments.

Sentiment Trend: The line plot tracks the average sentiment for each year, helping to analyze if content descriptions have become more positive or negative over time.

 Feature 4: Keyword Analysis in Descriptions

 What this cell does:

Identifies the most common keywords in the descriptions of movies and shows.

Helps in understanding what themes, topics, or genres are most frequent in the Netflix content.

Uses TF-IDF (Term Frequency-Inverse Document Frequency) to identify important keywords that stand out in the dataset.

```python
!pip install nltk
!pip install scikit-learn

import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
import pandas as pd
import numpy as np

# Download NLTK stopwords
nltk.download('stopwords')
stop_words = stopwords.words('english')
```

```python
# Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words=stop_words, max_features=50)

# Apply the vectorizer to the descriptions
tfidf_matrix = vectorizer.fit_transform(df['description'].dropna())

# Convert the TF-IDF matrix to a DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Display the top 10 keywords
top_keywords = tfidf_df.sum(axis=0).sort_values(ascending=False).head(10)
top_keywords
```

```python
# Plot the top 10 keywords as a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_keywords.values, y=top_keywords.index, palette='viridis')
plt.title('Top 10 Keywords in Netflix Descriptions')
plt.xlabel('TF-IDF Score')
plt.ylabel('Keyword')
plt.grid(True)
plt.show()
```

 Explanation:

TF-IDF: The TF-IDF score measures how important a word is in a particular document relative to all the documents in the dataset. The higher the score, the more important the word is in that document.

Stopwords: We remove common words like the, is, on because they don't provide meaningful insights into the content.

Top Keywords: By summing the TF-IDF scores across all the descriptions, we identify which keywords are most associated with Netflix's content.
